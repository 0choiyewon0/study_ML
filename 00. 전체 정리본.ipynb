# <span style="color:white; background-color:orange"> 03-2 경사 하강법에 대해 알아봅니다</span>

# <span style="color:blue; background-color:orange"> 선형 회귀</span>

1. y = ax + b 

    x(입력), y(타깃), a(기울기), b(절편)     
    
    ====================> 방정식 찾았어!
    
    =========>
     y_hat = wx +b
   
    x(입력), y_hat(예측값), w(가중치), b(절편)
    
    
    샘플을 근사한 방정식에 넣어 예측값 도출
    
    
 2. 경사하강법 : 모델이 데이터를 잘 표현할 수 있도록 기울기(변화율)을 사용하여 모델을 조금씩 조정하는 최적화 알고리즘
     
     많은 양의 데이터에 사용하기 좋은 알고리즘
     
     유일한 알고리즘이 아님을 주의 : 회귀 문제를 푸는 알고리즘은 정규 방정식, 결정 트리, 서포트 벡터 머신 등 여러가지 존재
     

3. 예측값 : 모델에 대해 새로운 입력값을 넣으면 어떤 출력이 나오는 데 이 값이 모델을 통해 예측한 값


4. 훈련 데이터에 잘 맞는 w,b 찾기
    

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. 가중치와 절편의 초기화

2. x[0] : 당뇨병 데이터에서의 3번째 특성 -> 첫번째([0]) 샘플(실제 데이터)

3. w,b 임의로 처음에 정한 값

4. 위의 설정값으로 y_hat 도출

5. y_hat과 실제 값인 y[0]를 서로 비교

```python
w = 1.0
b = 1.0
```

```python
y_hat = x[0] * w + b
print(y_hat)
```

```python
print(y[0])
```

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. w값을 조절하면서 예측값을 변경

2. 값을 +0.1로 조정하며 예측값과 실제값의 차이를 비교 


```python
w_inc = w + 0.1
y_hat_inc = w_inc * x[0] + b
print(y_hat_inc)
```

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. 조정한 w값의 크기에 따른 예측값의 변화율을 보기 위해 가중치 변화율 구하기

2. 구한 가중치의 변화율을 통하여 가중치 없데이트 공식에 사용

    가중치 변화율이 입력값임을 알 수 있음

```python
w_rate = (y_hat_inc - y_hat) / (w_inc - w)
print(w_rate)
```

```python
w_new = w + w_rate
print(w_new)
```

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. 가중치 조정값과 유사하게 절편 변화율을 구하는 과정

2. 절편의 변화율을 구해보면 1임을 알 수 있음

```python
b_inc = b + 0.1
y_hat_inc = x[0] * w + b_inc
print(y_hat_inc)
```

```python
b_rate = (y_hat_inc - y_hat) / (b_inc - b)
print(b_rate)
```

```python
b_new = b + 1
print(b_new)
```

# <span style="color:blue; background-color:orange"> 오차 역전파(backpropagation)</span>

1. 정의 : 예측값과 실제 타깃값의 차이를 이용하여 가중치와 절편을 업데이트 하는 과정


2. 오차 = 타깃값 - 예측값


3. 가중치와 절편을 업데이트할 때에 각각의 변화율에 err를 곱함


4. 이를 가지고 모든 샘플을 사용해 가중치와 절편을 업데이트 진행

```python
err = y[0] - y_hat
w_new = w + w_rate * err
b_new = b + 1 * err
print(w_new, b_new)
```

```python
y_hat = x[1] * w_new + b_new
err = y[1] - y_hat
w_rate = x[1]
w_new = w_new + w_rate * err
b_new = b_new + 1 * err
print(w_new, b_new)
```

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. 전체 샘플을 반복

2. zip(*iterale) : 동일한 개수로 이루어진 자료형을 묶어줌

3. 해당 과정을 완료 한 후 plt를 이용하여 산점도 출력

4. plot() 에 x좌표와 y좌표의 시작점과 종료점을 전달

    x좌표는 시작점 (-0.1) 종료점 (0.15) 를 넣고 y좌표는 w와 b를 사용하여 계산 

```python
for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b
    err = y_i - y_hat
    w_rate = x_i
    w = w + w_rate * err
    b = b + 1 * err
print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# <span style="color:blue; background-color:orange"> 아래 코드 설명</span>

1. 여러 에포크 반복

2. 에포크(epoch) : 전체 훈련 데이터를 모두 이용하여 한 단위의 작업을 진행하는 것

    일반적으로 수십에서 수춘 번의 에포크를 반복
    

```python
for i in range(1, 100):
    for x_i, y_i in zip(x, y):
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + 1 * err
print(w, b)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
class Neuron:
    
    '''
    __init__
    
    - 초기화 작업 수행
    - 가중치, 절편을 직접 초기화
    - __ : 파이썬이 자동으로 호출해 주는 메서드
    - self : 첫번째 인자
    - def 함수명(입력)
    
        return 반환
    - init안에서 self에 무언가를 대입하면 인스턴스에 속성이 추가
    - 점연선자 => 속성 바인딩(attribute binding) 수행하며 
                속성에 접근하면 먼저 인스턴스 검사하고 
                해당 속성이 없으면 인스턴스의 클래스를 계속 검사
    - self.속성값 : 프로퍼티 / self : 인스턴스 자기 자신을 의미
    '''
    def __init__(self):
        self.w = 1.0     
        self.b = 1.0     
    
    
      
    '''
    forpass
    
    - 직선 방정식을 계산(= 정방향 계산)
    - 예측값 = 입력값 * 가중치 + 절편
    '''
    def forpass(self, x):
        y_hat = x * self.w + self.b      
        return y_hat
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    '''
    def backprop(self, x, err):
        w_grad = x * err    
        b_grad = 1 * err  
        return w_grad, b_grad

    
    
    '''
    fit
    
    - 에포크 횟수 지정 : 값 지정 안하면 기본값 100으로 실행
    - 모든 샘플에 대해 반복
    - 흐름
        1. 정방향 계산
        2. 오차 계산
        3. 역방향 계산
        4. 가중치 업데이트
        5. 절편 업데이트
    '''
    def fit(self, x, y, epochs=100):
        for i in range(epochs):           
            for x_i, y_i in zip(x, y):    
                y_hat = self.forpass(x_i) 
                err = -(y_i - y_hat)     
                w_grad, b_grad = self.backprop(x_i, err)  
                self.w -= w_grad          
                self.b -= b_grad         
```

```python
neuron = Neuron()
neuron.fit(x, y)
```

```python
plt.scatter(x, y)
pt1 = (-0.1, -0.1 * neuron.w + neuron.b)
pt2 = (0.15, 0.15 * neuron.w + neuron.b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
class LogisticNeuron:
    
    
    
    '''
    __init__
    
    - 초기화 작업 수행
    - 절편과 가중치의 값을 따로 초기화 하지 않음
    - 특성 갯수에 따른 가중치 및 절편의 값이 유동적이게 되므로 번거롭지 않게 설정
    '''
    def __init__(self):
        self.w = None
        self.b = None

        
        
    '''
    forpass
    
    - 정방향 계산 수행
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    '''
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b 
        return z

    
    
    '''
    backprop
    
    - 역방향 계산
    - 가중치와 절편에 대한 그래디언트 계산 후 반환
    '''
    def backprop(self, x, err):
        w_grad = x * err    
        b_grad = 1 * err  
        return w_grad, b_grad

    
    
    '''
    activation
    
    - 활정화 함수 구현
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  
         max 값 보다 큰 값들을 max값으로 바꿈    
    '''
    def activation(self, z):
        z = np.clip(z, -100, None) 
        a = 1 / (1 + np.exp(-z)) 
        return a
        
        
        
    '''
    fit
    
    - 훈련 수행
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움    
    - np.zero() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 0로 채움    
    - np.full((num1,num2), num3) : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 num3로 채움    

    - 흐름
        1. 가중치와 절편 초기화
        2. epochs만큼 반복
        3. 모든 샘플에 대해 반복
        4. 정방향 계산
        5. 활성화 함수 적용
        6. 오차 계산
        7. 역방향 계산
        8. 가중치, 절편 업데이트
    '''
    def fit(self, x, y, epochs=100):
        self.w = np.ones(x.shape[1])     
        self.b = 0                       
        for i in range(epochs):          
            for x_i, y_i in zip(x, y):   
                z = self.forpass(x_i)   
                a = self.activation(z)    
                err = -(y_i - a)         
                w_grad, b_grad = self.backprop(x_i, err) 
                self.w -= w_grad         
                self.b -= b_grad          
    
    
    
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 정방향 계산 후 활성화 함수 적용
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    '''
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]   
        a = self.activation(np.array(z))       
        return a > 0.5
```

# <span style="color:blue; background-color:yellow">매 에포크마다 훈련 세트 샘플 순서 섞기</span>

1. 훈련 세트의 샘플 순서를 섞으면 가중치 최적값의 탐색 과정이 다양해져 가중치 최적값을 저대로 찾을 수 없음


2. 넘파이 배열의 인덱스를 섞은 후 인덱스 순서대로 샘플을 뽑는 것

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트 
    '''
    def __init__(self):
        self.w = None
        self.b = None
        self.losses = []

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    ''' 
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b 
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    '''   
    def backprop(self, x, err):
        w_grad = x * err    
        b_grad = 1 * err    
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  
         max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None) 
        a = 1 / (1 + np.exp(-z)) 
        return a
          
    
    
    '''
    fit
    
    - 훈련 수행
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
    - .shape( ) : 행렬의 차원 
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. epochs만큼 반복
        3. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        4. 정방향 계산
        5. 활성화 함수 적용
        6. 오차 계산
        7. 역방향 계산
        8. 가중치, 절편 업데이트

        9. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        10. 에포크 마다 평균 선실함수를 저장
    '''   
    def fit(self, x, y, epochs=100):
        self.w = np.ones(x.shape[1])              
        self.b = 0                               
        for i in range(epochs):                    
            loss = 0
            indexes = np.random.permutation(np.arange(len(x)))
            for i in indexes:                     
                z = self.forpass(x[i])             
                a = self.activation(z)             
                err = -(y[i] - a)               
                w_grad, b_grad = self.backprop(x[i], err)
                self.w -= w_grad                  
                self.b -= b_grad                  
              
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a)+(1-y[i])*np.log(1-a))
            
            self.losses.append(loss/len(y))
                
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    '''  
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]    
        return np.array(z) > 0                   
        
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
```

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음         
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None
        self.b = None
        self.losses = []
        self.val_losses = []
        self.w_history = []
        self.lr = learning_rate
        self.l1 = l1
        self.l2 = l2

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    '''        
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    '''   
    def backprop(self, x, err):
        w_grad = x * err          
        b_grad = 1 * err  
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  
         max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None) 
        a = 1 / (1 + np.exp(-z)) 
        return a
    
    
    
    '''
    fit
    
    - 훈련 수행
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
    - shape 속성 : 행렬의 차원 
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 역방향 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
            절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        self.w = np.ones(x.shape[1])               
        self.b = 0                                
        self.w_history.append(self.w.copy())      
        np.random.seed(42)
        for i in range(epochs):                    
            loss = 0
            indexes = np.random.permutation(np.arange(len(x)))
            for i in indexes:                     
                z = self.forpass(x[i])         
                a = self.activation(z)
                err = -(y[i] - a)                 
                w_grad, b_grad = self.backprop(x[i], err) 

                w_grad += self.l1 * np.sign(self.w) + self.l2 * self.w
                self.w -= self.lr * w_grad        
                self.b -= b_grad   
              
                self.w_history.append(self.w.copy())

                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a)+(1-y[i])*np.log(1-a))
                
            self.losses.append(loss/len(y) + self.reg_loss())
            self.update_val_loss(x_val, y_val)
            
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    '''  
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]     
        return np.array(z) >= 0                 
    
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
    
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        if x_val is None:
            return
        val_loss = 0
        for i in range(len(x_val)):
            z = self.forpass(x_val[i])     
            a = self.activation(z)         
            a = np.clip(a, 1e-10, 1-1e-10)
            val_loss += -(y_val[i]*np.log(a)+(1-y_val[i])*np.log(1-a))
        self.val_losses.append(val_loss/len(y_val) + self.reg_loss())
```

# <span style="color:blue; background-color:green">적절한 편향-분산 트레이드 오프</span>


1. 편향- 분산 트레이드오프(bias-variance tradeoff) : 과소 적합된 모델(펴냥)과 과대적합된 모델(분산)사이의 관계

    트레이드오프 : 하나를 얻기 위해서는 다른 하나를 희생해야 하기 때문
    
2. 편향을 줄임                  -> 분산 커짐

   (= 훈련 세트의 성능 높임)         (= 검증 세트와 성능 차이가 커짐)
   
   분산을 줄임                  -> 편향 커짐
   
   (= 검증 세트와 성능 차이가 줄임)    (= 훈련 세트의 선능 낮아짐)

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음         
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None
        self.b = None
        self.losses = []
        self.val_losses = []
        self.w_history = []
        self.lr = learning_rate
        self.l1 = l1
        self.l2 = l2

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    '''        
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    '''   
    def backprop(self, x, err):
        w_grad = x * err          
        b_grad = 1 * err  
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None) 
        a = 1 / (1 + np.exp(-z)) 
        return a
    
    
    
    '''
    fit
    
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
    - shape 속성 : 행렬의 차원 
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 역방향 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
            절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        self.w = np.ones(x.shape[1])               
        self.b = 0                                
        self.w_history.append(self.w.copy())      
        np.random.seed(42)
        for i in range(epochs):                    
            loss = 0
            indexes = np.random.permutation(np.arange(len(x)))
            for i in indexes:                     
                z = self.forpass(x[i])         
                a = self.activation(z)
                err = -(y[i] - a)                 
                w_grad, b_grad = self.backprop(x[i], err) 

                w_grad += self.l1 * np.sign(self.w) + self.l2 * self.w
                self.w -= self.lr * w_grad        
                self.b -= b_grad   
              
                self.w_history.append(self.w.copy())

                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a)+(1-y[i])*np.log(1-a))
  
            self.losses.append(loss/len(y) + self.reg_loss())
            self.update_val_loss(x_val, y_val)
            
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    '''  
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]     
        return np.array(z) >= 0                 
    
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
    
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        if x_val is None:
            return
        val_loss = 0
        for i in range(len(x_val)):
            z = self.forpass(x_val[i])     
            a = self.activation(z)        
            a = np.clip(a, 1e-10, 1-1e-10)
            val_loss += -(y_val[i]*np.log(a)+(1-y_val[i])*np.log(1-a))
        self.val_losses.append(val_loss/len(y_val) + self.reg_loss())
```

# <span style="color:blue; background-color:green">규제(regularization)</span>

0. 가중치 규제 : 가중치의 값이 커지지 않도록 제한하는 기법
                -> 모델의 일반화 성능이 올라감
                

    그래프를 통해서 본 규제


    회귀모델에서는 2번 모델이 더 적합: 기울기가 작을수록 일반화가 좋다
                                (기중릭가 작다 = 가중치가 낮아져 너무 작으면 과소 적합 모델)
    
    모델이 몇 개의 데이터에 집착
        -> 모델이 일반화 되지 않았다 : 새로운 데이터에 적응하지 못함(= 좋은 성능 아님)
        -> 규제를 사용하여 가중치를 제한
        -> 모델이 몇 개의 데이터에 집착하지 않게 됨
        -> 성능을 높일 수 있음
        
        
        
1. L1 규제(손실함수 + L1노름(norm))(선형회귀 + L1규제 = 라쏘(Lasso))

    
    - 정의 ; 가중치의 절댓값을 손실 함수에 더한 것
    - L1 노름의 n : 가중치의 개수
    - 로지스틱 손실함수 + (L1 노름)a
    
        a= 규제의 양을 조절하는 파라미터, 하이퍼파라미터(설정값)
        
        규제가 강해졌다 : a값이 크면 -> 전체 손실 함수의 값이 커지지 않도록 w 값의 합이 작아져야 함
        
        규제가 약해졌다 : a값이 작으면 -> w값이 커져도 손실 함수의 값이 큰 폭으로 커지지 않음
        
        
        가중치가 커지면 손실함수의 값 증가 -> 가중치 값 커지지 않게 노력
        (경사 하강법 = 손실함수를 낮추는 것)
        
     
     - 라쏘 : 가중치를 0으로 제작 가능
             -> 모델에서 사용할 수 없다
             -> 특성을 선택하는 효과(제외하는 효과)
             -> 모델의 복잡도가 떨어짐
             -> 사이킷런 : sklearn.linear_model.Lasso 라쏘 모델 제공
             
     - w_grad += alpha * np.sign(w) 
                
                sign(w) : 부호 값을 나타냄
                np.sign(w) : 배열 요소의 부호 반환이 이루어짐
                alpha 변수 : 규제 하이퍼파라미터 
                
                절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
                
                SGDClassifier클래스에서는 penalty 매개변수 값을 l1으로 지정하는 방법으로 L1 규제 적용 & alpha 매개변수 제공
             
==> L1 규제는 alph에 많이 의존
    

2. L2 규제(손실함수 + L2노름(norm))(선형회귀 + L2규제 = 릿지(Ridge))

    
    - 손실 함수에 가중치에 대한 L2 노름의 제곱을 더함
    - L2 노름 = 유클리드 거리
    - 로지스틱 손실함수 + (L2 노름)a
    - w_grad += alpha*w
      L2규제는 그레디언트 계산에 가중치 값 자체가 포함되므로 가중치의 부호만 사용하는 L1보다 조금 더 효과적 
      
    - 가중치를 완전히 0으로 만들지 않음
        
==> L1 규제보다 가중치가 큰 영향
    
    
3. 동시 사용 가능 규제의 강도는 alpha매개변수로 제어
     L1, L2 규제 동시 적용
                  L1 규제값 0 => L2만 진행
                  L2 규제값 0 => L1만 진행
                  
     선형회귀 + L1/L2 : 엘라스틱 넷(ElasticNet)

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음   
            
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None
        self.b = None
        self.losses = []
        self.val_losses = []
        self.w_history = []
        self.lr = learning_rate
        self.l1 = l1
        self.l2 = l2

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    '''        
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    '''   
    def backprop(self, x, err):
        w_grad = x * err          
        b_grad = 1 * err  
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None) 
        a = 1 / (1 + np.exp(-z)) 
        return a
    
    
    
    '''
    fit
    
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
    - shape 속성 : 행렬의 차원 
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 역방향 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
            경사 하강법으로 가중치를 업데이트하기 위해 규제를 적용한 로지스틱 손실 함수를 미분하여 아래 식 도출
                w_grad += alpha * np.sign(w) 
                
                sign(w) : 부호 값을 나타냄
                np.sign(w) : 배열 요소의 부호 반환이 이루어짐
                alpha 변수 : 규제 하이퍼파라미터 
                
                절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
                
                SGDClassifier클래스에서는 penalty 매개변수 값을 l1으로 지정하는 방법으로 L1 규제 적용 & alpha 매개변수 제공
                
                ---------
                w_grad += alpha*w
                
                L2규제는 그레디언트 계산에 가중치 값 자체가 포함되므로 가중치의 부호만 사용하는 L1보다 조금 더 효과적
                
                ----------
                L1, L2 규제 동시 적용
                  L1 규제값 0 => L2만 진행
                  L2 규제값 0 => L1만 진행
                  
                선형회귀 + L1/L2 : 엘라스틱 넷(ElasticNet)
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        self.w = np.ones(x.shape[1])               
        self.b = 0                                
        self.w_history.append(self.w.copy())      
        np.random.seed(42)
        for i in range(epochs):                    
            loss = 0
            indexes = np.random.permutation(np.arange(len(x)))
            for i in indexes:                     
                z = self.forpass(x[i])         
                a = self.activation(z)
                err = -(y[i] - a)                 
                w_grad, b_grad = self.backprop(x[i], err) 

                w_grad += self.l1 * np.sign(self.w) + self.l2 * self.w
                self.w -= self.lr * w_grad        
                self.b -= b_grad   
              
                self.w_history.append(self.w.copy())

                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a)+(1-y[i])*np.log(1-a))
  
            self.losses.append(loss/len(y) + self.reg_loss())
            self.update_val_loss(x_val, y_val)
            
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    '''  
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]     
        return np.array(z) >= 0                 
    
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산시 페널티 항에 대한 값을 더해야 함
    - 훈련 세트의 로지스틱 손실 함수값과 검증 세트의 로지스틱 손실 함수 값을 계산할 때 모두 호출
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
    
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        if x_val is None:
            return
        val_loss = 0
        for i in range(len(x_val)):
            z = self.forpass(x_val[i])     
            a = self.activation(z)        
            a = np.clip(a, 1e-10, 1-1e-10)
            val_loss += -(y_val[i]*np.log(a)+(1-y_val[i])*np.log(1-a))
        self.val_losses.append(val_loss/len(y_val) + self.reg_loss())
```

# <span style="color:blue; background-color:green">교차 검증(cross validation)</span>

0. 언제 사용 ?
    전체 데이터 세트의 샘플 개수가 많지 않을 때
    검증 세트를 훈련 세트에서 분리하느라 훈련 세트의 샘플 개수가 줄어들어 모델을 훈련 시킬 데이터가 부족해지는 경우 

         
        
1. 교차 검증 원리

    - 훈련 세트를 작은 덩어리로 나누어 진행
    - 작은 덩어리 : 폴드
        (five-fold cross vlidation)
    - 전체 Data가 검증 세트에 포함 -> 전체적인 검증 세트 점수가 안정적인 값이라고 판다 

2. 교차 검증 과정

    - 훈련 세트를 k개의 폴드(fold)로 나눔
    - 첫 번째 폴드를 검증 세트로 사용하고 나머지 폴드(k-1)를 훈련 세트로 사용
    - 모델을 훈련한 다음에 검증 세트로 평가
    - 차례대로 다음 폴드를 검증 세트로 사용하여 반복
    - k개의 검증 세트로 k번 성능을 평가한 후 계산된 성능의 평균을 내어 최종 성능을 계산
    
    
3. 교차 검증 훈련 세트 : k개의 폴드로 나누는 특징 (= k-폴드 교차 검증)
    
    - 모든 훈련 세트가 평가에 1번씩 사용됨
    - 검증 점수가 안정적
    - 기존의 훈련 방법보다 더 많은 데이터로 훈련 가능

# <span style="color:blue; background-color:green">아래 코드 설명</span>

1. 기존의 방식과 다르게 k-폴드 교차 검증에서는 검증 세트가 훈련 세트에 포함
    따라서 전체 데이터 세트를 다시 훈련 세트와 테스트 세트로 1번만 나눈 x_train_all, y_train_all을 훈련과 검증에 사용
    
2. 각 폴드의 검증 점수를 저장하기 위한 validation_scores 리스트 정의
    해당 값들을 평균하여 최정 검증 점수 계산

```python
validation_scores = []
```

# <span style="color:blue; background-color:green">아래 코드 설명</span>

1. k를 10으로 지정

2. bins 
    한개의 폴드에 들어갈 샘플의 개수 = 전체 훈련 세트의 샘플 개수/k

3. bins 변수의 개수만큼 건너뛰며 검증 폴드와 훈련 폴드를 구분하여 모델을 훈련

4. Start , end : 검증 폴드 샘플의 시작과 끝 인덱스 
    
5. train_index : list()함수를 이용하여 훈련 폴드의 인덱스를 모아둠

6. 모아둔 훈련 폴드 샘플의 인덱스로 train_fold와 train_target을 제작

7. 핵심 ) 
    - 훈련 데이터의 표준화 전처리를 폴드를 나눈 후에 수행
    - 폴드를 나누기 전에 전체 훈련 데이터를 전처리하면, 검증 폴드의 정보를 누설


```python
k = 10
bins = len(x_train_all) // k

for i in range(k):
    start = i*bins
    end = (i+1)*bins
    val_fold = x_train_all[start:end]
    val_target = y_train_all[start:end]
    
    train_index = list(range(0, start))+list(range(end, len(x_train_all)))
    train_fold = x_train_all[train_index]
    train_target = y_train_all[train_index]
    
    train_mean = np.mean(train_fold, axis=0)
    train_std = np.std(train_fold, axis=0)
    train_fold_scaled = (train_fold - train_mean) / train_std
    val_fold_scaled = (val_fold - train_mean) / train_std
    
    lyr = SingleLayer(l2=0.01)
    lyr.fit(train_fold_scaled, train_target, epochs=50)
    score = lyr.score(val_fold_scaled, val_target)
    validation_scores.append(score)

print(np.mean(validation_scores))
```

    배치 크기를 작게하여 처리하는 방식
    
    활률적 경사 하강법과 배치 경사 하강법의 절충안 방식

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음              
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None              
        self.b = None             
        self.losses = []           
        self.val_losses = []       
        self.w_history = []       
        self.lr = learning_rate    
        self.l1 = l1              
        self.l2 = l2               

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    - np.dot() : numpy array를 곱할 때 사용
    
    - 전장에서는 np.sum( ) 사용 하였는데, 이는 샘플이 1개씩 들어가기 때문에 .dot 사용
    '''      
    def forpass(self, x):
        z = np.dot(x, self.w) + self.b       
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    - 전체 샘플 수로 나누는 이유
        그레이디언트 평균값을 계산하기 위해
    '''   
    def backprop(self, x, err):
        m = len(x)
        w_grad = np.dot(x.T, err) / m        
        b_grad = np.sum(err) / m            
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None)            
        a = 1 / (1 + np.exp(-z))        
        return a
    
    
    
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - m : 샘플의 갯수 저장
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
        x.shape[1] : x의 2번째 차원을 보고(열의 갯수)
    - w_history : 가중치 변동시 저장
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
    - numpy.clip(array, min, max)
        array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 오차를 역전파하여 그레이디언트를 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
            경사 하강법으로 가중치를 업데이트하기 위해 규제를 적용한 로지스틱 손실 함수를 미분하여 아래 식 도출
                w_grad += alpha * np.sign(w) 
                
                sign(w) : 부호 값을 나타냄
                np.sign(w) : 배열 요소의 부호 반환이 이루어짐
                alpha 변수 : 규제 하이퍼파라미터 
                
                절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
                
                SGDClassifier클래스에서는 penalty 매개변수 값을 l1으로 지정하는 방법으로 L1 규제 적용 & alpha 매개변수 제공
                
                ---------
                w_grad += alpha*w
                
                L2규제는 그레디언트 계산에 가중치 값 자체가 포함되므로 가중치의 부호만 사용하는 L1보다 조금 더 효과적
                
                ----------
                L1, L2 규제 동시 적용
                  L1 규제값 0 => L2만 진행
                  L2 규제값 0 => L1만 진행
                  
                선형회귀 + L1/L2 : 엘라스틱 넷(ElasticNet)
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
        
        
    - 평균을 구할 때 샘플 갯수인 m으로 항상 나눔
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y = y.reshape(-1, 1)                  
        y_val = y_val.reshape(-1, 1)
        m = len(x)                            
        self.w = np.ones((x.shape[1], 1))
        self.b = 0                         
        self.w_history.append(self.w.copy())
      
        for i in range(epochs):
            z = self.forpass(x)              
            a = self.activation(z)
            err = -(y - a)              
         
            w_grad, b_grad = self.backprop(x, err)
        
            w_grad += (self.l1 * np.sign(self.w) + self.l2 * self.w) / m
      
            self.w -= self.lr * w_grad
            self.b -= self.lr * b_grad
 
            self.w_history.append(self.w.copy())
            a = np.clip(a, 1e-10, 1-1e-10)
            loss = np.sum(-(y*np.log(a) + (1-y)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / m)
            self.update_val_loss(x_val, y_val)
                
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    ''' 
    def predict(self, x):
        z = self.forpass(x)
        return z > 0            
        
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    - ch05에서는 y와 비교, 현재 장에서는 y.reshape(-1,1)와 비교
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y.reshape(-1, 1))
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산시 페널티 항에 대한 값을 더해야 함
    - 훈련 세트의 로지스틱 손실 함수값과 검증 세트의 로지스틱 손실 함수 값을 계산할 때 모두 호출
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
        
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        z = self.forpass(x_val)         
        a = self.activation(z)             
        a = np.clip(a, 1e-10, 1-1e-10)  
        val_loss = np.sum(-(y_val*np.log(a) + (1-y_val)*np.log(1-a)))
        self.val_losses.append((val_loss + self.reg_loss()) / len(y_val))
```

# <span style="color:blue; background-color:orange"> 오차 역전파(backpropagation)</span>

1. 정의 : 예측값과 실제 타깃값의 차이를 이용하여 가중치와 절편을 업데이트 하는 과정


2. 오차 = 타깃값 - 예측값


3. 가중치와 절편을 업데이트할 때에 각각의 변화율에 err를 곱함


4. 이를 가지고 모든 샘플을 사용해 가중치와 절편을 업데이트 진행

# <span style="color:yellow; background-color:blue"> 상속</span>

1. class 자식(부모) :
    상속받은 서브클래스에서는 상속해준 슈퍼 클래스의 속성과 메소드 모두 사용 가능
    
    자식에서 재정의가 가능하며, super을 통해 부모를 호출 

```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음              
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None              
        self.b = None             
        self.losses = []           
        self.val_losses = []       
        self.w_history = []       
        self.lr = learning_rate    
        self.l1 = l1              
        self.l2 = l2               

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    - np.dot() : numpy array를 곱할 때 사용
    
    - 전장에서는 np.sum( ) 사용 하였는데, 이는 샘플이 1개씩 들어가기 때문에 .dot 사용
    '''      
    def forpass(self, x):
        z = np.dot(x, self.w) + self.b       
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    - 전체 샘플 수로 나누는 이유
        그레이디언트 평균값을 계산하기 위해
    '''   
    def backprop(self, x, err):
        m = len(x)
        w_grad = np.dot(x.T, err) / m        
        b_grad = np.sum(err) / m            
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None)            
        a = 1 / (1 + np.exp(-z))        
        return a
    
    
    
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - m : 샘플의 갯수 저장
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
        x.shape[1] : x의 2번째 차원을 보고(열의 갯수)
    - w_history : 가중치 변동시 저장
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
       - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 오차를 역전파하여 그레이디언트를 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
            경사 하강법으로 가중치를 업데이트하기 위해 규제를 적용한 로지스틱 손실 함수를 미분하여 아래 식 도출
                w_grad += alpha * np.sign(w) 
                
                sign(w) : 부호 값을 나타냄
                np.sign(w) : 배열 요소의 부호 반환이 이루어짐
                alpha 변수 : 규제 하이퍼파라미터 
                
                절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
                
                SGDClassifier클래스에서는 penalty 매개변수 값을 l1으로 지정하는 방법으로 L1 규제 적용 & alpha 매개변수 제공
                
                ---------
                w_grad += alpha*w
                
                L2규제는 그레디언트 계산에 가중치 값 자체가 포함되므로 가중치의 부호만 사용하는 L1보다 조금 더 효과적
                
                ----------
                L1, L2 규제 동시 적용
                  L1 규제값 0 => L2만 진행
                  L2 규제값 0 => L1만 진행
                  
                선형회귀 + L1/L2 : 엘라스틱 넷(ElasticNet)
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
        
        
    - 평균을 구할 때 샘플 갯수인 m으로 항상 나눔
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y = y.reshape(-1, 1)                  
        y_val = y_val.reshape(-1, 1)
        m = len(x)                            
        self.w = np.ones((x.shape[1], 1))
        self.b = 0                         
        self.w_history.append(self.w.copy())
      
        for i in range(epochs):
            z = self.forpass(x)              
            a = self.activation(z)
            err = -(y - a)              
         
            w_grad, b_grad = self.backprop(x, err)
        
            w_grad += (self.l1 * np.sign(self.w) + self.l2 * self.w) / m
      
            self.w -= self.lr * w_grad
            self.b -= self.lr * b_grad
 
            self.w_history.append(self.w.copy())
            a = np.clip(a, 1e-10, 1-1e-10)
            loss = np.sum(-(y*np.log(a) + (1-y)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / m)
            self.update_val_loss(x_val, y_val)
                
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    ''' 
    def predict(self, x):
        z = self.forpass(x)
        return z > 0            
        
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    - ch05에서는 y와 비교, 현재 장에서는 y.reshape(-1,1)와 비교
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y.reshape(-1, 1))
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산시 페널티 항에 대한 값을 더해야 함
    - 훈련 세트의 로지스틱 손실 함수값과 검증 세트의 로지스틱 손실 함수 값을 계산할 때 모두 호출
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
        
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        z = self.forpass(x_val)         
        a = self.activation(z)             
        a = np.clip(a, 1e-10, 1-1e-10)  
        val_loss = np.sum(-(y_val*np.log(a) + (1-y_val)*np.log(1-a)))
        self.val_losses.append((val_loss + self.reg_loss()) / len(y_val))
```

# <span style="color:yellow; background-color:blue"> backprop</span>



```python
class DualLayer(SingleLayer):
    
    
    
    
    '''
    __init__
    
    - losses : 손실 함수의 결과 값 저장할 리스트
    - 학습률 : 0.1 
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음              
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    - 흐름
        1. 은닉층의 뉴런 갯수
        2. 은닉층의 가중치
        3. 은닉층의 절편
        4. 출력층의 가중치
        5. 출력층의 절편
        6. 은닉층의 활성화값  => 역방향 계산시 필요
        7. 훈련 손실
        8. 검증 손실
        9. 학습률
        10. l1, l2
    '''
    def __init__(self, units=10, learning_rate=0.1, l1=0, l2=0):
        self.units = units       
        self.w1 = None            
        self.b1 = None            
        self.w2 = None            
        self.b2 = None            
        self.a1 = None           
        self.losses = []       
        self.val_losses = []      
        self.lr = learning_rate   
        self.l1 = l1               
        self.l2 = l2               
        
        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    - np.dot() : numpy array를 곱할 때 사용   => 첫 번째 층의 선형 식 계산
    - 활성화 함수 적용
    - 활성화 함수의 결과값을 가지고 마치 입력값처럼 두 번쨰 층의 선형 식을 계산
    '''    
    def forpass(self, x):
        z1 = np.dot(x, self.w1) + self.b1       
        self.a1 = self.activation(z1)
        z2 = np.dot(self.a1, self.w2) + self.b2  
        return z2
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - m : 샘플의 개수
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    - 전체 샘플 수로 나누는 이유
        그레이디언트 평균값을 계산하기 위해  ===> 샘플 개수에 영향이 없음
    - err_to_hidden : 시그모이드 함수까지 그래디언트 계산
    - axis=매개변수 
        =none  : 모든 요소
        =1     : (@, #, O)->(#, O)
        =2     : (@, #, O)->(@, O)
        =3     : (@, #, O)->(@, #)
        
    '''  
    def backprop(self, x, err):
        m = len(x)      
        w2_grad = np.dot(self.a1.T, err) / m
        b2_grad = np.sum(err) / m
        err_to_hidden = np.dot(err, self.w2.T) * self.a1 * (1 - self.a1)
        # 은닉층의 가중치와 절편에 대한 그래디언트를 계산합니다.
        w1_grad = np.dot(x.T, err_to_hidden) / m
        b1_grad = np.sum(err_to_hidden, axis=0) / m
        return w1_grad, b1_grad, w2_grad, b2_grad
    
    
    
    '''
    init_weights
    
    - 앞서 fit매서드에 있던 가중치 초기화 부분을 분리 한 것
    - 가중치 초기화가 중요하기 때문에
    - 전체 코드 짧게 하기 위해서
    - w1, b1은 은닉층 관련 코드
    - w2, b2는 출력층 관련 코드
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움    
    - np.zero() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 0로 채움    
    - np.full((num1,num2), num3) : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 num3로 채움    
    
    '''
    def init_weights(self, n_features):
        self.w1 = np.ones((n_features, self.units))  # (특성 개수, 은닉층의 크기)
        self.b1 = np.zeros(self.units)               # 은닉층의 크기
        self.w2 = np.ones((self.units, 1))           # (은닉층의 크기, 1)
        self.b2 = 0
    
    
    
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - m : 샘플의 갯수 저장
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움 ==> init_weights로 뺌
        x.shape[1] : x의 2번째 차원을 보고(열의 갯수)
    - numpy.clip(array, min, max)
        array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    - 평균을 구할 때 샘플 갯수인 m으로 항상 나눔

    - 흐름
        1. 타깃을 열 벡터로 변경
        2. 검증 타깃을 열 벡터로 변경
        3. 샘플의 개수 저장
        4. 은닉층과 출력층의 가중치를 초기화
        5. 에포크 만큼 반복
        6. 안전한 로그 계산을 위한 clip 사용
        7. 로그 손실과 규제 손실을 더하여 리스트에 추가
        8. 검증 세트에 대한 손실을 계산
        
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y = y.reshape(-1, 1)          
        y_val = y_val.reshape(-1, 1)
        m = len(x)                    
        self.init_weights(x.shape[1]) 
        for i in range(epochs):
            a = self.training(x, y, m)
            a = np.clip(a, 1e-10, 1-1e-10)
            loss = np.sum(-(y*np.log(a) + (1-y)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / m)
            self.update_val_loss(x_val, y_val)
      
    
    
    
    '''
    training
    
    - 정방향 계산과 그레이디언트 업데이트 코드만 뺌
    - 훈련 데이터 x,y와 훈련 샘플의 개수 m을 매개 변수로 받고 마지막 출력층의 활성화출력 a반환
    - 흐름
        1. 정방향 계산 수행
        2. 활성화 함수 적용
        3. 오차를 역전파하여 그레이디언트 계산
        4. 그레이디언트에서 페널티 항의 미분값을 더함
        5. 은닉층의 가중치와 절편을 업데이트
        6. 출력층의 가중치와 절편을 업데이트
    '''
    def training(self, x, y, m):
        z = self.forpass(x)      
        a = self.activation(z)  
        err = -(y - a)            
        w1_grad, b1_grad, w2_grad, b2_grad = self.backprop(x, err)
        w1_grad += (self.l1 * np.sign(self.w1) + self.l2 * self.w1) / m
        w2_grad += (self.l1 * np.sign(self.w2) + self.l2 * self.w2) / m
        self.w1 -= self.lr * w1_grad
        self.b1 -= self.lr * b1_grad
        self.w2 -= self.lr * w2_grad
        self.b2 -= self.lr * b2_grad
        return a
    
    
    
    
    
    '''
    reg_loss
    
    - 은닉층과 출력층의 가중치에 대한 l1, l2손실을 계산
    '''
    def reg_loss(self):
        return self.l1 * (np.sum(np.abs(self.w1)) + np.sum(np.abs(self.w2))) + \
               self.l2 / 2 * (np.sum(self.w1**2) + np.sum(self.w2**2))
```

# <span style="color:yellow; background-color:blue">아래 코드 설명</span>

1. 가중치 초기화 개선
    
    손실 그래프를 보니 초기 손실값이 감소하는 곡선이 매끄럽지 못함
    
    손실 함수가 감소하는 방향을 올바르게 찾는데 시간이 많이 소요
    
    지금까지 가중치 1로 놓고 훈련을 시작했으나 random.normal( )을 이용하여 초기화
    
    
    
2. random.normal( ) 
    
    정규 분포를 다르는 무작위 수로 가중치를 초기화
    

```python
class RandomInitNetwork(DualLayer):
    
    
    '''
    init_weights
    
    - 층이 많아 질수록 초기값 시작점이 중요하다
    - 복잡도 높아질수록
    - 엉뚱한 곳에서 시작하거나 지역 최저점에 빠질수도 있음
    - random.seed( ) 함수를 통하여 무작위 수의 초기값을 고정
    - np.random.normla(평균, 표준편차, 배열크기)
    
    '''
    def init_weights(self, n_features):
        np.random.seed(42)
        self.w1 = np.random.normal(0, 1, 
                                   (n_features, self.units))  # (특성 개수, 은닉층의 크기)
        self.b1 = np.zeros(self.units)                        # 은닉층의 크기
        self.w2 = np.random.normal(0, 1, 
                                   (self.units, 1))           # (은닉층의 크기, 1)
        self.b2 = 0
```

코드(Code) 영역
우리가 작성한 소스코드가 들어 가는 부분. 즉, 실행할 프로그램의 코드가 저장되는 영역으로 텍스트(code)영역이라고도 부른다. 코드영역은 실행 파일을 구성하는 명령어들이 올라가는 메모리 영역으로 함수, 제어문, 상수 등이 여기에 지정된다.


데이터(Data) 영역
전역변수와 static변수가 할당되는 영역

스택(Stack) 영역
프로그램이 자동으로 사용하는 임시 메모리 영역이다.

힙(Heap) 영역
프로그래머가 할당/해제하는 메모리 공간이다.
Java에서는 가비지 컬렉터가 자동으로 해제합니다.
이 공간에 메모리 할당하는 것을 동적 할당(Dynamic Memory Allocation)이라고도 부른다.


```python
class SingleLayer:
    
    '''
    __init__
    
    - 초기화 작업을 수행
    - 앞선 뉴런 클래스에서의 __init__와 유사
    - 이제 가중치와 절편의 값(w,b)를 직접 지정하지 않음
    - losses : 손실 함수의 결과 값 저장할 리스트
    - w_history : 가중치 변동시 저장할 리스트
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음              
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    '''
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None              
        self.b = None             
        self.losses = []           
        self.val_losses = []       
        self.w_history = []       
        self.lr = learning_rate    
        self.l1 = l1              
        self.l2 = l2               

        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    - np.dot() : numpy array를 곱할 때 사용
    
    - 전장에서는 np.sum( ) 사용 하였는데, 이는 샘플이 1개씩 들어가기 때문에 .dot 사용
    '''      
    def forpass(self, x):
        z = np.dot(x, self.w) + self.b       
        return z
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    - 전체 샘플 수로 나누는 이유
        그레이디언트 평균값을 계산하기 위해
    '''   
    def backprop(self, x, err):
        m = len(x)
        w_grad = np.dot(x.T, err) / m        
        b_grad = np.sum(err) / m            
        return w_grad, b_grad
    
    
    
    '''
    activation
    
    - 시그모이드 함수 사용
    - np.exp() 함수는 자연 상수의 지수 함수를 계산
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    '''  
    def activation(self, z):
        z = np.clip(z, -100, None)            
        a = 1 / (1 + np.exp(-z))        
        return a
    
    
    
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - m : 샘플의 갯수 저장
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움
        x.shape[1] : x의 2번째 차원을 보고(열의 갯수)
    - w_history : 가중치 변동시 저장
    - .copy() : 얕은 사본
      .deepcopy() : 깊은 사본
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
    - np.sign : 배열 원소의 부호 판단
       - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    - 흐름
        1. 가중치와 절편을 각각 초기화
        2. 가중치의 변화에 따라 w_history리스트에 가중치를 기록
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. epochs만큼 반복
        5. index값을 무작위로 선언
            : 훈련 세트의 샘플 순서를 섞는 전형적인 방법 
              훈련 세트 자체를 섞는 것보다 효율적이고 빠름
        6. 정방향 계산
        7. 활성화 함수 적용
        8. 오차 계산
        9. 오차를 역전파하여 그레이디언트를 계산
        10. 가중치의 그래디언트에 l1, l2의 변화값을 더해줌
            경사 하강법으로 가중치를 업데이트하기 위해 규제를 적용한 로지스틱 손실 함수를 미분하여 아래 식 도출
                w_grad += alpha * np.sign(w) 
                
                sign(w) : 부호 값을 나타냄
                np.sign(w) : 배열 요소의 부호 반환이 이루어짐
                alpha 변수 : 규제 하이퍼파라미터 
                
                절편의 변화량은 조절하지 않는 이유는 이는 변화에 영향을 미치지 않고 이동에만 영향을 줌
                
                SGDClassifier클래스에서는 penalty 매개변수 값을 l1으로 지정하는 방법으로 L1 규제 적용 & alpha 매개변수 제공
                
                ---------
                w_grad += alpha*w
                
                L2규제는 그레디언트 계산에 가중치 값 자체가 포함되므로 가중치의 부호만 사용하는 L1보다 조금 더 효과적
                
                ----------
                L1, L2 규제 동시 적용
                  L1 규제값 0 => L2만 진행
                  L2 규제값 0 => L1만 진행
                  
                선형회귀 + L1/L2 : 엘라스틱 넷(ElasticNet)
        11. 가중치의 변화량 조절을 위해 lr의 값을 곱함
        12. 변화 된 가중치를 w_history 에 추가
        13. 안전한 로그 계산을 위해 clip() 함수 적용 후 손실을 누적 (시그모이드 사용시 안전하게)
        14. 에포크 마다 평균 선실함수를 저장
        15. 검증 세트에 대한 손실을 계산 
        
        
    - 평균을 구할 때 샘플 갯수인 m으로 항상 나눔
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y = y.reshape(-1, 1)                  
        y_val = y_val.reshape(-1, 1)
        m = len(x)                            
        self.w = np.ones((x.shape[1], 1))
        self.b = 0                         
        self.w_history.append(self.w.copy())
      
        for i in range(epochs):
            z = self.forpass(x)              
            a = self.activation(z)
            err = -(y - a)              
         
            w_grad, b_grad = self.backprop(x, err)
        
            w_grad += (self.l1 * np.sign(self.w) + self.l2 * self.w) / m
      
            self.w -= self.lr * w_grad
            self.b -= self.lr * b_grad
 
            self.w_history.append(self.w.copy())
            a = np.clip(a, 1e-10, 1-1e-10)
            loss = np.sum(-(y*np.log(a) + (1-y)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / m)
            self.update_val_loss(x_val, y_val)
                
            
            
    '''
    predict
    
    - 훈련된 데이터를 가지고 예측 값을 계산
    - 예측값 :  입력값을 선형 함수, 활성화 함수, 임계 함수 순서로 통과시 구현
    - z의 계산 : 리스트 내포(list comprehension)
         ([])안에 for문을 삽입하여 새 리스트 생성
         x의 행을 하나씩 꺼낸 수 forpass() 메서드 적용 후 결과를 이용하여 새 리스트로 생성
    - 리스트 z를 넘파이 배열로 변환(np.array) 후 activation()에 전달
    - a>0.5 : 양수   (= z>0)
      a<0.5 : 음수   (= z<0)
    ''' 
    def predict(self, x):
        z = self.forpass(x)
        return z > 0            
        
    
    
    '''
    score
    
    - test세트, 훈련 세트의 데이터를 타깃값과 함께 넣어 예측 결과 값이 얼마나 나오는지를 평가(= 정확도 측정)
    - np.mean() : true=1 , false =0
    - ch05에서는 y와 비교, 현재 장에서는 y.reshape(-1,1)와 비교
    ''' 
    def score(self, x, y):
        return np.mean(self.predict(x) == y.reshape(-1, 1))
    
    
    
    '''
    reg_loss

    - 로지스틱 손실 함수의 계산시 페널티 항에 대한 값을 더해야 함
    - 훈련 세트의 로지스틱 손실 함수값과 검증 세트의 로지스틱 손실 함수 값을 계산할 때 모두 호출
    ''' 
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
        
    
    
    '''
    update_val_loss
    
    - 검증 세트의 손실점수의 값 업데이트
    - 정방향 계산과 활성화 함수 적용을 반복적으로 진행
    - np.clip을 통해 시그모이드 함수를 안전하게 진행
    ''' 
    def update_val_loss(self, x_val, y_val):
        z = self.forpass(x_val)         
        a = self.activation(z)             
        a = np.clip(a, 1e-10, 1-1e-10)  
        val_loss = np.sum(-(y_val*np.log(a) + (1-y_val)*np.log(1-a)))
        self.val_losses.append((val_loss + self.reg_loss()) / len(y_val))
```

```python
class DualLayer(SingleLayer):
    
    
    
    
    '''
    __init__
    
    - losses : 손실 함수의 결과 값 저장할 리스트
    - 학습률 : 0.1 
    - lr (learning_rate)
        학습률을 의미하며, 이 값으로 가중치의 업데이트 양을 조절
        왜? 손실 함수는 복잡한 굴곡을 가진 다차원 공간의 초평면(hyperplane)
            큰 폭으로 가중치를 업데이트 하다보면 최적의 해(최적의 가중치와 절편)을 구할 수 없음              
    - l1, l2 : L1, L2 규제
        l1 : L1 규제의 alpha 값
        l2 : L2 규제의 alpha 값
        기본값은 0 , 이때 규제를 적용하지 않음
    - 흐름
        1. 은닉층의 뉴런 갯수
        2. 은닉층의 가중치
        3. 은닉층의 절편
        4. 출력층의 가중치
        5. 출력층의 절편
        6. 은닉층의 활성화값  => 역방향 계산시 필요
        7. 훈련 손실
        8. 검증 손실
        9. 학습률
        10. l1, l2
    '''
    def __init__(self, units=10, learning_rate=0.1, l1=0, l2=0):
        self.units = units       
        self.w1 = None            
        self.b1 = None            
        self.w2 = None            
        self.b2 = None            
        self.a1 = None           
        self.losses = []       
        self.val_losses = []      
        self.lr = learning_rate   
        self.l1 = l1               
        self.l2 = l2               
        
        
        
    '''
    forpass
    
    - 직선 방정식을 계산
    - 정방향 계산
    - np.sum() : 배열의 같은 위치를 더해주며, 크기 값은 변화하지 않음  
    - np.dot() : numpy array를 곱할 때 사용   => 첫 번째 층의 선형 식 계산
    - 활성화 함수 적용
    - 활성화 함수의 결과값을 가지고 마치 입력값처럼 두 번쨰 층의 선형 식을 계산
    '''    
    def forpass(self, x):
        z1 = np.dot(x, self.w1) + self.b1       
        self.a1 = self.activation(z1)
        z2 = np.dot(self.a1, self.w2) + self.b2  
        return z2
    
    
    
    '''
    backprop
    
    - 가중치와 절편에 대한 그래디언트 계산
    - 역방향 계산
    - m : 샘플의 개수
    - err와 함께 곱하는 값들은 앞서 계산한 가중치와 절편의 기울기
        가중치 => 입력값
        절편 => 1
    - 전체 샘플 수로 나누는 이유
        그레이디언트 평균값을 계산하기 위해  ===> 샘플 개수에 영향이 없음
    - err_to_hidden : 시그모이드 함수까지 그래디언트 계산
    - axis=매개변수 
        =none  : 모든 요소
        =1     : (@, #, O)->(#, O)
        =2     : (@, #, O)->(@, O)
        =3     : (@, #, O)->(@, #)
        
    '''  
    def backprop(self, x, err):
        m = len(x)      
        w2_grad = np.dot(self.a1.T, err) / m
        b2_grad = np.sum(err) / m
        err_to_hidden = np.dot(err, self.w2.T) * self.a1 * (1 - self.a1)
        # 은닉층의 가중치와 절편에 대한 그래디언트를 계산합니다.
        w1_grad = np.dot(x.T, err_to_hidden) / m
        b1_grad = np.sum(err_to_hidden, axis=0) / m
        return w1_grad, b1_grad, w2_grad, b2_grad
    
    
    
    '''
    init_weights
    
    - 앞서 fit매서드에 있던 가중치 초기화 부분을 분리 한 것
    - 가중치 초기화가 중요하기 때문에
    - 전체 코드 짧게 하기 위해서
    - w1, b1은 은닉층 관련 코드
    - w2, b2는 출력층 관련 코드
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움    
    - np.zero() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 0로 채움    
    - np.full((num1,num2), num3) : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 num3로 채움    
    
    '''
    def init_weights(self, n_features):
        self.w1 = np.ones((n_features, self.units))  # (특성 개수, 은닉층의 크기)
        self.b1 = np.zeros(self.units)               # 은닉층의 크기
        self.w2 = np.ones((self.units, 1))           # (은닉층의 크기, 1)
        self.b2 = 0
    
    
    
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - m : 샘플의 갯수 저장
    - np.ones() : 입력된 매개변수와 동일한 크기의 배열을 만들고 값을 모두 1로 채움 ==> init_weights로 뺌
        x.shape[1] : x의 2번째 차원을 보고(열의 갯수)
    - numpy.clip(array, min, max)
        array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  max 값 보다 큰 값들을 max값으로 바꿈       
    - 평균을 구할 때 샘플 갯수인 m으로 항상 나눔

    - 흐름
        1. 타깃을 열 벡터로 변경
        2. 검증 타깃을 열 벡터로 변경
        3. 샘플의 개수 저장
        4. 은닉층과 출력층의 가중치를 초기화
        5. 에포크 만큼 반복
        6. 안전한 로그 계산을 위한 clip 사용
        7. 로그 손실과 규제 손실을 더하여 리스트에 추가
        8. 검증 세트에 대한 손실을 계산
        
    ''' 
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y = y.reshape(-1, 1)          
        y_val = y_val.reshape(-1, 1)
        m = len(x)                    
        self.init_weights(x.shape[1]) 
        for i in range(epochs):
            a = self.training(x, y, m)
            a = np.clip(a, 1e-10, 1-1e-10)
            loss = np.sum(-(y*np.log(a) + (1-y)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / m)
            self.update_val_loss(x_val, y_val)
      
    
    
    
    '''
    training
    
    - 정방향 계산과 그레이디언트 업데이트 코드만 뺌
    - 훈련 데이터 x,y와 훈련 샘플의 개수 m을 매개 변수로 받고 마지막 출력층의 활성화출력 a반환
    - 흐름
        1. 정방향 계산 수행
        2. 활성화 함수 적용
        3. 오차를 역전파하여 그레이디언트 계산
        4. 그레이디언트에서 페널티 항의 미분값을 더함
        5. 은닉층의 가중치와 절편을 업데이트
        6. 출력층의 가중치와 절편을 업데이트
    '''
    def training(self, x, y, m):
        z = self.forpass(x)      
        a = self.activation(z)  
        err = -(y - a)            
        w1_grad, b1_grad, w2_grad, b2_grad = self.backprop(x, err)
        w1_grad += (self.l1 * np.sign(self.w1) + self.l2 * self.w1) / m
        w2_grad += (self.l1 * np.sign(self.w2) + self.l2 * self.w2) / m
        self.w1 -= self.lr * w1_grad
        self.b1 -= self.lr * b1_grad
        self.w2 -= self.lr * w2_grad
        self.b2 -= self.lr * b2_grad
        return a
    
    
    
    
    
    '''
    reg_loss
    
    - 은닉층과 출력층의 가중치에 대한 l1, l2손실을 계산
    '''
    def reg_loss(self):
        return self.l1 * (np.sum(np.abs(self.w1)) + np.sum(np.abs(self.w2))) + \
               self.l2 / 2 * (np.sum(self.w1**2) + np.sum(self.w2**2))
```

```python
class RandomInitNetwork(DualLayer):
    
    
    '''
    init_weights
    
    - 층이 많아 질수록 초기값 시작점이 중요하다
    - 복잡도 높아질수록
    - 엉뚱한 곳에서 시작하거나 지역 최저점에 빠질수도 있음
    - random.seed( ) 함수를 통하여 무작위 수의 초기값을 고정
    - np.random.normla(평균, 표준편차, 배열크기)
    
    '''
    def init_weights(self, n_features):
        np.random.seed(42)
        self.w1 = np.random.normal(0, 1, 
                                   (n_features, self.units))  # (특성 개수, 은닉층의 크기)
        self.b1 = np.zeros(self.units)                        # 은닉층의 크기
        self.w2 = np.random.normal(0, 1, 
                                   (self.units, 1))           # (은닉층의 크기, 1)
        self.b2 = 0
```

```python
class MinibatchNetwork(RandomInitNetwork):
        
    
    
    
    '''
    __init__
    
    - batch_size 매개 변수 직접 관리 
    - 가장 좋은 것을 찾아야 함 : 2의 배수 : 계산 효율성이 좋음
    '''
    def __init__(self, units=10, batch_size=32, learning_rate=0.1, l1=0, l2=0):
        super().__init__(units, learning_rate, l1, l2)
        self.batch_size = batch_size    
        
        
        
    '''
    fit
    
    - .reshape 속성 : (변경할 배열, 차원), 배열.reshape(차원) 
    - .reshape(-1,1) : 전치 행렬
    - numpy.clip(array, min, max)
         array 내의 element들에 대해서 min 값 보다 작은 값들을 min값으로 바꿔주고  
         max 값 보다 큰 값들을 max값으로 바꿈    
         
    - init_weights
        - 층이 많아 질수록 초기값 시작점이 중요하다
        - 복잡도 높아질수록
        - 엉뚱한 곳에서 시작하거나 지역 최저점에 빠질수도 있음
        - random.seed( ) 함수를 통하여 무작위 수의 초기값을 고정
        - np.random.normla(평균, 표준편차, 배열크기)
        
        
    - training
    
        - 정방향 계산과 그레이디언트 업데이트 코드만 뺌
        - 훈련 데이터 x,y와 훈련 샘플의 개수 m을 매개 변수로 받고 마지막 출력층의 활성화출력 a반환
        - 흐름
            1. 정방향 계산 수행
            2. 활성화 함수 적용
            3. 오차를 역전파하여 그레이디언트 계산
            4. 그레이디언트에서 페널티 항의 미분값을 더함
            5. 은닉층의 가중치와 절편을 업데이트
            6. 출력층의 가중치와 절편을 업데이트
            
    - 흐름
        1. 검증 타깃을 열 벡터로 변경
        2. 은닉층과 출력층의 가중치를 초기화
        3. 랜덤 시드를 지정(fit 함수에서는 42)
        4. 에포크 만큼 반복
        5. 두 번째 for문에서 손실을 누적하기 위해 에포크마다 loss 변수 초기화
        6. DualLayer과 다르게 샘플 개수m을 미니배치 마다 계산
        7. gen_batch에서 반환한 미니배치를 순환
        8. 타깃을 열 벡터로 변경
        9. 샘플 개수를 저장
        10. 반환된 미니 배치 데이터를 training( )메서드에게 전달
        11. 안전한 로그 계산을 위한 클리핑
        12 로그 손실과 규제 손실을 더해 리스트에 추가
        13 검증 세트에 대한 손실을 계산
    '''
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        y_val = y_val.reshape(-1, 1)    
        self.init_weights(x.shape[1])    
        np.random.seed(42)
        for i in range(epochs):
            loss = 0
            for x_batch, y_batch in self.gen_batch(x, y):
                y_batch = y_batch.reshape(-1, 1) 
                m = len(x_batch)                
                a = self.training(x_batch, y_batch, m)
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += np.sum(-(y_batch*np.log(a) + (1-y_batch)*np.log(1-a)))
            self.losses.append((loss + self.reg_loss()) / len(x))
            self.update_val_loss(x_val, y_val)


            
            
            
    '''
    gen_batch
    
    - 파이썬 제너레이터(generator)로 구현
        순차적으로 데이터에 접근할 수 있는 반복 가능한 객체(iterator)를 반환
        명시적으로 리스트를 만들지 않으면서 필요한 만큼 데이터를 추출할 수 있음  ==> 효율적인 메모리 사용 가능
    - np.random.permutation()
        무작위로 섞인 배열 생성
        (숫자) : 해당 숫자 까지의 숫자를 가진 배열 생성
        (np.arange(len(x))) : 입력값의 길이 크기를 범위로 지정
        
        => 인덱스를 섞으면서 전체 훈련 샘플을 섞는 효과를 줌

    - batch_size 만큼씩 x,y 배열을 건너뛰며 미니 배치를 반환
    - gen_batch 메서드가 호출될 때마다(에포크만큼) 훈련 데이터 배열의 인덱스를 섞음
    '''
    def gen_batch(self, x, y):
        length = len(x)
        bins = length // self.batch_size # 미니배치 횟수
        if length % self.batch_size:
            bins += 1                    # 나누어 떨어지지 않을 때
        indexes = np.random.permutation(np.arange(len(x))) # 인덱스를 섞습니다.
        x = x[indexes]
        y = y[indexes]
        for i in range(bins):
            start = self.batch_size * i
            end = self.batch_size * (i + 1)
            yield x[start:end], y[start:end]   # batch_size만큼 슬라이싱하여 반환합니다.
```
